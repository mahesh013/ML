```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"D:\Machine Learning Practical\Position_Salaries.csv")
```

```python
print("Dataset preview:\n", data.head())
```

```python
X = data[['Level']].values  # Independent variable
y = data['Salary'].values 
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
levels_to_predict = np.array([[11], [12]])
predicted_salaries = model.predict(levels_to_predict)
```

```python
print("\nPredicted salaries:")
for level, salary in zip(levels_to_predict.flatten(), predicted_salaries):
    print(f"Level {level}: {salary:.2f}")
```

```python
y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)
print(f"\nMean Squared Error of the model: {mse:.2f}")
```

```python
# Visualize the results
plt.scatter(X, y, color='blue', label="Actual Data")
plt.plot(X, y_pred, color='red', label="Regression Line")
plt.scatter(levels_to_predict, predicted_salaries, color='green', label="Predicted Levels", zorder=5)
plt.xlabel("Level")
plt.ylabel("Salary")
plt.title("Simple Linear Regression (Level vs Salary)")
plt.legend()
plt.show()
```
